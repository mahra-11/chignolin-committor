# Notebook 2 — Committor Probability Counting

For each microstate cluster, estimate the committor probability q(i):  
the fraction of trajectory visits to cluster i that next reach the  
**folded state (A)** before the **unfolded state (B)**.

**Strategy ("counting once"):**  
Consecutive frames in the same cluster are collapsed into one visit.  
For each visit we scan forward in the trajectory until hitting A or B,  
and count the outcome. This avoids overcounting from frames where the  
system lingers in a cluster for many steps.

**Input:** `data/<system>/cluster_<N>.csv`  
**Output:** `data/<system>/cluster_<N>_prob.csv` (adds `Committor_prob` column)

```python
import yaml
import sys
sys.path.insert(0, "..")

from src.committor import compute_committor, write_committor_csv, plot_committor
```

```python
# --- Select system and cluster file ---
with open("../config/chignolin.yaml") as f:   # change to ww_domain.yaml for WW domain
    cfg = yaml.safe_load(f)

SYSTEM = cfg["system"]
FOLDED   = cfg["folded_clusters"]    # e.g. [1]
UNFOLDED = cfg["unfolded_clusters"]  # e.g. [7]

N = 40   # cluster count — change to match the CSV you want to annotate
CLUSTER_CSV = f"../data/{SYSTEM}/cluster_{N}.csv"
OUT_CSV     = f"../data/{SYSTEM}/cluster_{N}_prob.csv"

print(f"System:   {SYSTEM}")
print(f"Folded:   {FOLDED}")
print(f"Unfolded: {UNFOLDED}")
print(f"Input:    {CLUSTER_CSV}")
```

```python
# Compute committor probabilities
committor_dict = compute_committor(CLUSTER_CSV, FOLDED, UNFOLDED)
print(f"Clusters with committor data: {len(committor_dict)}")
```

```python
# Visualize committor distribution
plot_committor(committor_dict, title=f"{SYSTEM} — {N} clusters")
```

```python
# Write annotated CSV
write_committor_csv(CLUSTER_CSV, OUT_CSV, committor_dict, FOLDED, UNFOLDED)
```
